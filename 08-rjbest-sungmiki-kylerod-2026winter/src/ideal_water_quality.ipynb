# Information regarding ideal water quality

#### **About good / bad Notebook**

There are 6 new columns (DRP, EC, Alkalinity, how_good, ideal & unideal).

```DRP``` = 1 if value of ```dissolved reactive phosphorus``` is in the ideal range, else 0,  
```EC``` = 1 if value of ```electrical conductance``` is in the ideal range, else 0,  
```Alkalinity``` = 1 if value of ```total alkalinity``` is in the ideal range, else 0,  
```how_good``` = adding up the three previous columns. Possible values are 1, 2 or 3 (if all of the previous values are 1, then 3).  

```ideal``` = 1 if how_good is 3, else 0,  
```unideal``` = 1 if how_good is 0.  

If Alkalinity, EC and DRP are all outside the ideal water quality range, then there is a 1 in the bad column, otherwise there is a 0.

#### Load Dependencies and Load Data

````python
import pandas as pd
import numpy as np

# read water quality data
df = pd.read_csv('data/wq.csv')
df.describe().round().astype(int)
````

#### When the primary dataset was acquired, information was provided about the 3 key water quality parameters, and their ideal range of values.

Total Alkalinity - Ideal Range: 20 to 200 mg/L  -> milligrams per liter or ppm  
Electrical Conductance - Ideal Range: < 800 µS/cm -> microsiemens per centimeter  
Dissolved Reactive Phosphorus - Ideal Range: < 100 µg/L -> micrograms per liter

````python
# Add a column for each key value, and label it 1 if the instances value is within the ideal range, and 0 otherwise.
df['DRP'] = [1 if _ < 100 else 0 for _ in df['dissolved reactive phosphorus']]
df['EC'] = [1 if _ < 800 else 0 for _ in df['electrical conductance']]
df['Alkalinity'] = [1 if _ > 20 and _ < 200 else 0 for _ in df['total alkalinity']]

# check for nulls
df.isnull().sum()
````

````python
# Here are DataFrames with only ideal or undeal water quality
ideal = df[(df['DRP']==1) & (df['EC']==1) & (df['Alkalinity']==1)]
unideal = df[(df['DRP']==0) & (df['EC']==0) & (df['Alkalinity']==0)]

# Add ideal/unideal columns
df = df.assign(how_good=df['Alkalinity']+df['DRP']+df['EC'])
df['ideal'] = np.where(df['how_good'] > 2, 1, 0)
df['unideal'] = np.where(df['how_good'] < 1, 1, 0)
df.head()
````
